```python
import random

def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:
        comparisons += 1

        # Avoid division by zero
        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            break

        pos = low + int(
            ((key - arr[low]) * (high - low)) /
            (arr[high] - arr[low])
        )

        if arr[pos] == key:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Dataset sizes
sizes = [10000, 50000, 100000]

for n in sizes:
    # Generate uniformly distributed floating-point numbers
    arr = sorted(random.uniform(0.0, 1000.0) for _ in range(n))

    # Choose an existing element to ensure successful search
    key = arr[n // 2]

    index, comparisons = interpolation_search(arr, key)

    print("Dataset Size:", n)
    print("Element Found at Index:", index)
    print("Number of Comparisons:", comparisons)
    print("-" * 35)
```
